```python
# Hassan Ali

# Import libraries 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Loading dataset from CSV file
bal_df = pd.read_csv('balloons.csv')
y = bal_df['inflated']  # Target variable
```

```python
# Plotting bar graph showing the distribution of the target variable
plt.bar([1, 2], [len(y[y == 'T']), len(y[y == 'F'])], tick_label=['T', 'F'])
plt.show()  
```

```python
# Converting categorical data to integer representations for all columns in the dataframe
# Necessary as Random Forest in sklearn requires numerical input
for col in bal_df.columns:
    labels, uniques = pd.factorize(bal_df[col])
    bal_df[col] = labels
```

```python
# Updating target variable 'y' after conversion and preparing features 'X' by dropping the target column
y = bal_df['inflated']
X = bal_df.drop(columns='inflated')
```

```python
# Spliting the dataset into training and testing sets. 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Initializing Random Forest classifier with 500 trees and fit it to the training data
rfc = RandomForestClassifier(n_estimators=500)
rfc.fit(X_train, y_train)
```

```python
# Make predictions on the test set and printing accuracy of the model on the test set
predictions = rfc.predict(X_test)
print("Accuracy=", accuracy_score(y_test, predictions))
```

```python
# Printing confusion matrix to evaluate the model's performance in more detail
print(confusion_matrix(y_test, predictions))
```
