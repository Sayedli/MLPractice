```python
dis = df['dis']
nox = df['nox']

dis_np = dis.to_numpy()


# 2. `bedrooms_np = bedrooms.to_numpy()`: Similar to the first line, this converts another Pandas Series (`bedrooms`, which represent the number of bedrooms in a
# housing dataset) into a NumPy array.
nox_np = nox.to_numpy()

# 3. `pop_np.shape, bedrooms_np.shape`: This line evaluates the shape of both NumPy arrays created from the Series. The `.shape` attribute of a NumPy array returns a
# tuple representing the dimensions of the array. If the Series were one-dimensional (which is typical for Series), their shapes would be `(n,)`, where `n` is the
# number of elements in the Series.
print(dis_np.shape, nox_np.shape)
```

```python
from sklearn.linear_model import LinearRegression

# 2. `sklearn_model = LinearRegression().fit(pop_np.reshape((17000, 1)), bedrooms_np)`: This line creates an instance of `LinearRegression` and fits it to the data.
# The `fit` method is used to fit the linear model to the data:
#   2.1 - `pop_np.reshape((17000, 1))` suggests that there is a NumPy array named `pop_np` which is being reshaped into a two-dimensional array with 17,000 rows and 1 column.
#   This represents the population feature from the dataset, which is being used as the independent variable or predictor.
#   2.1 - `bedrooms_np` is a NumPy array representing the number of bedrooms in each instance of the dataset, used as the dependent variable or target.
# The `.fit()` method is what actually trains the linear regression model. The model is trying to find the best linear relationship between the population size
# (as the independent variable) and the number of bedrooms (as the dependent variable).


model = LinearRegression()
model.fit(dis_np.reshape(-1, 1), nox_np)
```

```python
predicted_nox = model.predict(dis_np.reshape(-1, 1))

plt.scatter(dis, nox, label='Actual Data')
plt.scatter(dis, predicted_nox, color='red', label='Predicted Data')
plt.xlabel('Dis')
plt.ylabel('Total Nox')
plt.title('Linear Regression Predictions')
plt.legend()
plt.show()
```

```python
# Define the predictor variable 'X' (independent variable) and the response variable 'y' (dependent variable)
X = df['dis']  # The predictor variable is the column 'x' from the dataset
y = df['nox']  # The response variable is the column 'y' from the dataset

# Add a constant term to 'X' to account for the intercept in the linear regression model
# This step ensures the model can calculate both the slope and the intercept
X = sm.add_constant(X)

# Fit a linear regression model using the Ordinary Least Squares (OLS) method
# The 'fit()' method estimates the model parameters (slope and intercept) based on the data
lr_model = sm.OLS(y, X).fit()

# Plotting the regression results

# Label the x-axis with the name of the predictor variable
plt.xlabel('Dis')

# Label the y-axis with the name of the response variable
plt.ylabel('Nox')

# Set the title of the plot to describe what the visualization represents
plt.title('Linear Regression (OLS)')

# Define a range of tick marks for the x-axis, starting from 0 and incrementing by 5
# The range extends slightly beyond the maximum value in the 'x' column for better visualization
xt = np.arange(0, df['dis'].max() + 0.1, 15)

# Define a range of tick marks for the y-axis, starting from 0 and incrementing by 10
# The range extends slightly beyond the maximum value in the 'y' column for better visualization
yt = np.arange(0, df['nox'].max() + 0.001, 0.7)

# Set the tick marks for the x-axis and y-axis using the defined ranges
plt.xticks(xt)
plt.yticks(yt)

# Set the limits for the x-axis and y-axis to control the range of data shown in the plot
plt.xlim(0, df['dis'].max() + 1)  # X-axis limits
plt.ylim(0, df['nox'].max() + 0.09)  # Y-axis limits

# Plot the data points from the dataset
# Each point represents a pair of values from the 'x' and 'y' columns
plt.scatter(df['dis'], df['nox'], color='blue')  # Data points are shown as blue dots

# Use the fitted regression model to predict the 'y' values based on 'X'
y_pred = lr_model.predict(X)

# Plot the regression line, which shows the predicted relationship between 'x' and 'y'
# The regression line is displayed in red for distinction
plt.plot(df['dis'], y_pred, color='red')

# Display the plot with the data points and the regression line
plt.show()
```

```python
# Prepare the data for regression analysis
X = df['lstat']  # Independent variable
y = df['medv']  # Dependent variable
X = sm.add_constant(X)  # Adds a constant term to the predictor

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the summary of the linear regression model
print(model.summary())
print()
print("Linear Reg: " + str(model.params['lstat']) + "*x  +" + str(model.params['const']))

# Plotting
plt.scatter(df['lstat'], df['medv'])
plt.xlabel('lstat')
plt.ylabel('medv')
plt.title('Scatter plot of lstat vs medv')

y_pred = model.predict(X)
plt.plot(df['lstat'], y_pred, color='red')  # Regression line in red
plt.show() 
```
