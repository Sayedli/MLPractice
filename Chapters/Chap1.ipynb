```python
# Exercise 1.1

# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Height and Weight data from Table 1.3
heights = np.array([61, 62, 63, 64, 65, 66, 67, 68, 69, 70, 71, 72]).reshape(-1, 1)
weights = np.array([122, 124, 125, 132, 145, 151, 157, 165, 172, 176, 178, 180])

# Creating Linear Regression model
model = LinearRegression()
model.fit(heights, weights)

# Predicting weights for given heights
predicted_weights = model.predict(np.array([60, 73, 65.5, 68.5]).reshape(-1, 1))

# Output predictions
print("Predicted Weights:")
print(f"60 inches: {predicted_weights[0]:.2f} lbs")
print(f"73 inches: {predicted_weights[1]:.2f} lbs")
print(f"65.5 inches: {predicted_weights[2]:.2f} lbs")
print(f"68.5 inches: {predicted_weights[3]:.2f} lbs")

# Visualizing data and the regression line
plt.figure(figsize=(10, 6))
plt.scatter(heights, weights, color='blue', label="Data Points")
plt.plot(heights, model.predict(heights), color='red', label="Regression Line")
plt.scatter([60, 73, 65.5, 68.5], predicted_weights, color='green', marker='x', label="Predicted Points")
plt.xlabel("Height (inches)")
plt.ylabel("Weight (lbs)")
plt.title("Height vs Weight with Linear Regression")
plt.legend()
plt.grid()
plt.show()
```

```python
# Exercise 1.2

# Importing libraries
import pandas as pd

# Data from Table 1.4
feed_data = {
    "Weight": [179, 160, 136, 227, 217, 168, 108, 124, 143, 140, 309, 229, 181, 
               141, 260, 203, 148, 169, 213, 257, 244, 271, 243, 230, 248, 327, 329],
    "Feed": ["Horsebean", "Horsebean", "Horsebean", "Horsebean", "Horsebean", "Horsebean", 
             "Horsebean", "Horsebean", "Horsebean", "Horsebean", "Linseed", "Linseed", 
             "Linseed", "Linseed", "Linseed", "Linseed", "Linseed", "Linseed", "Linseed", 
             "Linseed", "Linseed", "Linseed", "Soybean", "Soybean", "Soybean", "Soybean", "Soybean"]
}

# Creating DataFrame
df = pd.DataFrame(feed_data)

# Grouping by 'Feed' and calculate mean weights
mean_weights = df.groupby("Feed")["Weight"].mean()

# Finding the feed type with the highest average weight
best_feed = mean_weights.idxmax()
best_feed_weight = mean_weights.max()

# Display results 
print("Average Weight by Feed Type:")
formatted_mean_weights = mean_weights.reset_index()
print(formatted_mean_weights.to_string(index=False, header=["Feed", "Weight"]))

print(f"\nThe best feed type is '{best_feed}' with an average weight of {best_feed_weight:.2f} lbs.")
```

```python
# Exercise 1.3

# Importing libraries
import pandas as pd

# Data from Table 1.5
insurance_data = {
    "Insurance Provider": ["GEICO", "GEICO", "GEICO",
                           "Progressive", "Progressive", "Progressive",
                           "USAA", "USAA", "USAA"],
    "Rating": [4.7, 8.3, 9.2, 7.4, 6.7, 8.9, 3.8, 6.3, 8.1]
}

# Creating DataFrame
df = pd.DataFrame(insurance_data)

# Grouping by 'Insurance Provider' and calculate mean ratings
mean_ratings = df.groupby("Insurance Provider")["Rating"].mean()

# Finding the provider with the highest average rating
best_provider = mean_ratings.idxmax()
best_provider_rating = mean_ratings.max()

# Display results 
print("Average Ratings by Insurance Provider:")
formatted_ratings = mean_ratings.reset_index()
print(formatted_ratings.to_string(index=False, header=["Insurance Provider", "Average Rating"]))

print(f"\nThe best insurance provider is '{best_provider}' with an average rating of {best_provider_rating:.2f}.")
```
